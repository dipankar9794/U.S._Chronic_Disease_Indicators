```python
import pandas as pd

# Load dataset
df = pd.read_csv('C:/Users/Lenovo/Desktop/U.S._Chronic_Disease_Indicators.csv')

# Basic exploration
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt

# Count entries by year
year_counts = df['YearStart'].value_counts().sort_index()

# Plot
year_counts.plot(kind='bar', figsize=(10, 4), color='skyblue')
plt.title('Number of Records per Year')
plt.xlabel('Year')
plt.ylabel('Record Count')
plt.tight_layout()
plt.show()
```

```python
state_counts = df['LocationAbbr'].value_counts().head(10)

state_counts.plot(kind='bar', color='orange')
plt.title('Top 10 States with Most Data Entries')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate missing values by column
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent
}).sort_values(by='Missing %', ascending=False)

print(missing_summary.head(10))
```

```python
# Count data sources
data_sources = df['DataSource'].value_counts().head(10)

# Bar plot of top sources
data_sources.plot(kind='barh', color='mediumseagreen')
plt.title('Top 10 Data Sources')
plt.xlabel('Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
indicator_filter = df[df['Question'].str.contains('Obesity', case=False, na=False)]
gender_group = indicator_filter.groupby('Stratification1')['DataValue'].mean()

gender_group.plot(kind='bar', color=['lightblue', 'lightpink'])
plt.title('Average Obesity by Gender')
plt.ylabel('Average Data Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
state = 'CA'
year = 2019

subset = df[(df['LocationAbbr'] == state) & (df['YearStart'] == year)].dropna(subset=['LowConfidenceLimit', 'HighConfidenceLimit'])

subset['ErrorRange'] = subset['HighConfidenceLimit'] - subset['LowConfidenceLimit']
subset = subset.sort_values('ErrorRange', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.barh(subset['Question'], subset['ErrorRange'], color='coral')
plt.xlabel('Confidence Interval Range')
plt.title(f'Indicators with Highest Uncertainty in {state}, {year}')
plt.tight_layout()
plt.show()
```

```python
topic_counts = df['Topic'].value_counts().head(10)

plt.figure(figsize=(8, 8))
plt.pie(topic_counts, labels=topic_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Health Topics (by Record Count)')
plt.tight_layout()
plt.show()
```

```python
indicator = 'No leisure-time physical activity among adults'

gender_data = df[(df['Question'] == indicator) & df['Stratification1'].isin(['Male', 'Female'])]
gender_avg = gender_data.groupby('Stratification1')['DataValue'].mean()

gender_avg.plot(kind='bar', color=['steelblue', 'lightcoral'])
plt.title(f'{indicator} by Gender')
plt.ylabel('Average Data Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
diabetes = df[df['Question'].str.contains('diabetes', case=False, na=False)]
diabetes_avg = diabetes.groupby('LocationAbbr')['DataValue'].mean().sort_values(ascending=False).head(10)

diabetes_avg.plot(kind='bar', color='mediumpurple')
plt.title('Top 10 States with Highest Diabetes Prevalence')
plt.xlabel('State')
plt.ylabel('Average Data Value')
plt.tight_layout()
plt.show()
```

```python
missing_by_state = df.groupby('LocationAbbr').apply(lambda x: x['DataValue'].isnull().mean()).sort_values(ascending=False).head(10)

missing_by_state.plot(kind='barh', color='firebrick')
plt.title('Top 10 States with Highest Missing Data Ratio')
plt.xlabel('Proportion of Missing Data')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
